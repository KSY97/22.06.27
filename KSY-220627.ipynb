```python
# 2번 문제
pets = [
    {"name": "구름", "age" : 5},
    {"name": "초코", "age" : 3},
    {"name": "아지", "age" : 1},
    {"name": "호랑이", "age" : 1},
]

print("# 우리 동네 애완 동물들")
for dic in pets:
    print(f"{dic['name']} {dic['age']}살")
```

```python
# 3번 문제
numbers = [1,2,6,8,4,3,2,1,9,5,4,9,7,2,1,3,5,4,8,9,7,2,3]
counter = {}

for number in numbers:
    if counter.get(number) == None:
        counter[number] = 0
        
    for num in range(1,10):
        if num == number:
            counter[num] = counter.get(num) + 1
print(counter)
```

```python
# 4번 문제
char = {
    "name": "기사",
    "level": "12",
    "items": {
        "sword": "불꽃의 검",
        "armor": "풀플레이트"
    },
    "skill": ["베기", "세게 베기", "아주 세게 베기"]
}

for key in char:
    if type(char[key]) is list:
        char_3 = char[key]
        for key_3 in char_3:
            print(f"{key} : {key_3}")
    elif type(char[key]) is dict:
        char_2 = char[key]
        for key_2 in char_2:
            print(f"{key_2} : {char_2[key_2]}")
    else:
        print(f"{key} : {char[key]}")
```
